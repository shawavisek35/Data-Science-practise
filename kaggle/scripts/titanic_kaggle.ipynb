# Performing data wrangling

```python
#providing a discreate value for the sex column using the apply function
def make_sex_column_discreate(gender):
    if gender=="female":
        return 0
    else:
        return 1
    
titanic_train_data['Sex'] = titanic_train_data['Sex'].apply(make_sex_column_discreate)
titanic_train_data
```

```python
titanic_train_data.remove_columns(column_names=['Name' , 'Age' , 'Fare' , 'Ticket'] , inplace=True)
```

```python
def modify_embarked(s):
    if s=="C":
        return 0
    elif s=="S":
        return 1
    else:
        return 2
    
titanic_train_data['Embarked'] = titanic_train_data['Embarked'].apply(modify_embarked)
```

```python
titanic_train_data
```

```python
turicreate.visualization.item_frequency(titanic_train_data["Cabin"])
```

```python
#Creating logistic regression model
my_model = turicreate.logistic_classifier.create(titanic_train_data , target = "Survived" , features=['Pclass' , 'Sex' , 'SibSp' , 'Parch' , 'Embarked'])
```

```python
#Data Wrangling for the test set
titanic_test_data['Sex'] = titanic_test_data['Sex'].apply(make_sex_column_discreate)
```

```python
titanic_test_data['Embarked'] = titanic_test_data['Embarked'].apply(modify_embarked)
```

```python
#Evaluating the model
titanic_test_data['Survived'] = my_model.predict(titanic_test_data)
```

```python
titanic_test_data
```

```python
submission = titanic_test_data.remove_columns(column_names=['Pclass' , 'Name' , 'Sex' , 'Age' , 'SibSp' , 'Parch' , 'Ticket' , 'Fare' , 'Cabin' , 'Embarked'])
```
